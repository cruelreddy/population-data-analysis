```python
import pandas as pd
```

```python
df_population_raw = pd.read_csv('population_total.csv')


```

```python
df_population_raw.dropna(inplace = True)
df_population_raw1 = df_population_raw.pivot(index = 'year',columns = 'country',values = 'population')
df_population_raw1
```

```python
df_pivot = df_population_raw1
```

```python
df_pivot = df_pivot[['Afghanistan','Zimbabwe','Vietnam','Uzbekistan','India']]
```

```python
df_pivot
```

```python
#visualization
```

```python
#line 
```

```python
df_pivot.plot(kind = 'line',xlabel = 'year',title='populationyears',figsize = (10,5))
```
